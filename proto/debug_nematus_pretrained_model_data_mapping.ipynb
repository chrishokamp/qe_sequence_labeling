```python
import json
import os
import codecs
from collections import OrderedDict, Counter
```

```python
PRETRAINED_SYSTEM = '/media/1tb_drive/nematus_ape_experiments/amunmt_ape_pretrained/system/models/src-pe'

src_vocab = json.load(codecs.open(os.path.join(PRETRAINED_SYSTEM, 'vocab.src.json'), encoding='utf8'))
trg_vocab = json.load(codecs.open(os.path.join(PRETRAINED_SYSTEM, 'vocab.pe.json'), encoding='utf8'))

datadir = '/media/1tb_drive/Dropbox/data/qe/amunmt_artificial_ape_2016/data/concat_500k_with_wmt16/'

src_data = codecs.open(os.path.join(datadir, 'train.src.prepped'), encoding='utf8') 
src_tokens = [l.strip().split() for l in src_data] 

trg_data = codecs.open(os.path.join(datadir, 'train.mt.prepped'), encoding='utf8') 
trg_tokens = [l.strip().split() for l in src_data]
```

```python
def get_unks(tokens, vocab):
    unk_toks = Counter()
    for r in tokens:
        unk_toks.update([t for t in r if t not in vocab])
    return unk_toks
        
```

```python
unk_src = get_unks(src_tokens, src_vocab).most_common()
unk_trg = get_unks(trg_tokens, trg_vocab).most_common()
```
